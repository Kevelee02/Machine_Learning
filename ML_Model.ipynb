```python
import pandas as pd

melbourne_file_path = "https://raw.githubusercontent.com/dphi-official/First_ML_Model/master/melbourne_data.csv"
melbourne_data = pd.read_csv(melbourne_file_path) 
melbourne_data.columns
```

```python
#Definimos o nosso target ou alvo, que vai ser o pre√ßo
y = melbourne_data.Price
```

```python
#Agora definimos nossas features, as variaveis que iremos usar para prever o nosso target
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = melbourne_data[melbourne_features]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Dando uma olhada na correlacao
sns.heatmap(melbourne_data[melbourne_features].corr(), annot=True)
plt.show()
```

```python
#Agora iremos construir de fato o primeiro modelo

from sklearn.tree import DecisionTreeRegressor

# Definimos o modelo que iremos usar
melbourne_model = DecisionTreeRegressor(random_state=1)

# Definimos o dados
melbourne_model.fit(x, y)
```

```python
#Por fim olhamos os pre√ßos previstos
print(x.head())
print("Os pre√ßos previstos s√£o: ")
print(melbourne_model.predict(x.head()))

#Quantos mais features ou variaveis tivermos, melhor ir√° ser a previs√£o, mas esse foi um modelo simples de ML(meu primeiro)
```
